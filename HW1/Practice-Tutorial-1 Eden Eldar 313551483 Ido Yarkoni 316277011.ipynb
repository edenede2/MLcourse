Write a function named **cube_volume** that when given a length of a line, returns the area of a cube.

```python
def cube_volume(num):
    return num ** 3

# Tests
print(cube_volume(3))  # 27
print(cube_volume(5))  # 125
print(cube_volume(7))  # 343
```

## Assignment #2

Write a function named **parse_words**, that given a string prints each word in a different line.

<p> For example parse_words ("<em><strong>Lorem ipsum dolor sit amet, vis dolore soleat</em></strong>")  will pe printed as</p>
<p> Lorem </p>
<p> ipsum </p>
<p> dolor </p>
<p> sit </p>
<p> amet, </p>
<p> vis </p>
<p> dolore </p>
<p> soleat </p>

```python
def parse_words(sentence):
    sentence = sentence.split()
    for i in sentence:
        print(i)
# Tests
parse_words("I love Python!")  # I\nlove\nPython!
parse_words("Hello, world!")  # Hello,\nworld!
parse_words("This is a test.")  # This\nis\na\ntest.
```

<p> Write a function <strong> today_date </strong> that prints today's date in the format DD-MM-YYYY. </p>
<p> For example if today is March 10, 2023, the function will print <em> 10-03-2023 </em>. </p>

```python
from datetime import datetime

def today_date():
    return datetime.today().strftime('%d-%m-%Y')

# Tests
print(today_date()) # DD-MM-YYYY format
```
