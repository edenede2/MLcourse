```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
iris = pd.read_csv('iris.csv')
iris.rename(columns={'Species': 'variety',
                     'SepalLengthCm': 'sepal.length',
                     'SepalWidthCm': 'sepal.width',
                     'PetalLengthCm': 'petal.length',
                     'PetalWidthCm': 'petal.width'
                     }, inplace=True)
iris['variety'] = iris['variety'].apply(lambda x: x.split('-')[1])
iris['variety'] = iris['variety'].apply(lambda x: x.capitalize())
iris.drop(['Unnamed: 0', 'Unnamed: 0.1', 'Unnamed: 0.2', 'Id'], axis=1, inplace=True)
columns = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width', 'variety']
iris = iris[columns]
```

```python
iris_corr = iris[['petal.length', 'petal.width', 'sepal.length', 'sepal.width']].corr()
```

```python
display(iris_corr)
```

```python
plt.figure(figsize = (10,8))
ax = sns.heatmap(data=iris_corr, cmap= 'RdBu_r', vmin=-1, vmax=1, annot=True, annot_kws={"size": 10})

ax.yaxis.set_tick_params(labelsize=10)  # Set yticklabels size

plt.show()
```

```python
g = sns.pairplot(iris)
g.fig.set_size_inches(12,12)
plt.show()
```

```python
color_palette = sns.color_palette("hls", 3)
g = sns.pairplot(iris, hue='variety', corner=True, palette=color_palette)
g.fig.set_size_inches(12,12)
```

### Assignment #3

(a) Construct 3 different **heatmaps**, one for each iris specie.

This might require splitting the dataframe <strong> iris </strong> into 3 different pandas dataframes.

```python
iris_setosa = iris[iris['variety'] == 'Setosa'].drop('variety', axis=1)
iris_versicolor = iris[iris['variety'] == 'Versicolor'].drop('variety', axis=1)
iris_virginica = iris[iris['variety'] == 'Virginica'].drop('variety', axis=1)
```

Space for 3 heatmaps below:

```python
iris_setosa_corr = iris_setosa.corr()
plt.figure(figsize = (8,6))
sns.heatmap(data=iris_setosa_corr, cmap= 'RdBu_r', vmin=-1, vmax=1, annot=True, annot_kws={"size":10})
plt.title('Setosa')
plt.show()
```

```python
iris_versicolor_corr = iris_versicolor.corr()
plt.figure(figsize = (8,6))
sns.heatmap(data=iris_versicolor_corr, cmap= 'RdBu_r', vmin=-1, vmax=1, annot=True, annot_kws={"size":10})
plt.title('Versicolor')
plt.show()
```

```python
iris_virginica_corr = iris_virginica.corr()
plt.figure(figsize = (8,6))
sns.heatmap(data=iris_virginica_corr, cmap= 'RdBu_r', vmin=-1, vmax=1, annot=True, annot_kws={"size":10})
plt.title('Virginica')
plt.show()
```
